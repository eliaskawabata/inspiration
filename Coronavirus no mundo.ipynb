# 0.0 Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.signal import savgol_filter
%matplotlib inline
```

## 0.1 Data Load

```python
df_raw = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv')
```

```python
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
```

#  2.0 Location

```python
df2 = df1.copy()
```

## 2.1 India

```python
df_india = df2[df2['location'] == 'India']
```

```python
y_india = df_india['new_deaths_smoothed_per_million']
x_india = df_india['date']
plt.figure(figsize=(15,8))
plt.plot(x_india, y_india, color = 'red', label = 'India')
plt.title('Óbitos por milhão')
plt.ylabel('Óbitos por milhão')
plt.legend()
```
